```python
import pandas as pd
```

```python
met3 = pd.read_csv("raw_data\MeTDiff.v1_m3.csv", delimiter=";")
met4 = pd.read_csv("raw_data\MeTDiff.v1_m4.csv", delimiter=";")
met6 = pd.read_csv("raw_data\MeTDiff.v1_m6.csv", delimiter=";")
```
